```python
import pandas as pd
import numpy as np

import sklearn
from sklearn.feature_extraction.text import CountVectorizer

from sklearn import preprocessing
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import precision_recall_fscore_support
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix, plot_confusion_matrix, classification_report, accuracy_score

import time

from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('data/breastcancerwisconsin 16032021170504.csv')

df
```

```python
df = df.drop('Unnamed: 32', axis=1)

#descreve o dataset
df.describe()
```

```python
#converte o atributo diagnosis para numérico
df['diagnosis'] = pd.factorize(df['diagnosis'])[0]
df = pd.get_dummies(df)
df
```

```python
#puxa a matriz de correlação de todos os atributos numéricos
df.corr()
```

```python
def classifier(X_train,X_test,Y_train,Y_test): 
    #Aplicando KNN
    print(" KNN ")
    ini = time.time() #Inicio do calculo do tempo

    knn=KNeighborsClassifier()
    knn.fit(X_train,Y_train)

    end = time.time()
    knn_time = end - ini # Fim do calculo do tempo

    print(classification_report(Y_test,knn.predict(X_test)))
    print()

    #Aplicando Árvore de Decisão
    print("Árvore de Decisão")
    ini = time.time() #Inicio do calculo do tempo

    dtc = tree.DecisionTreeClassifier()
    dtc.fit(X_train, Y_train)

    end = time.time()
    dtc_time = end - ini # Fim do calculo do tempo

    print(classification_report(Y_test,dtc.predict(X_test)))

    return knn, dtc
```

```python
features = df[df.columns.drop(['diagnosis'])]._get_numeric_data()

X_train,X_test,Y_train,Y_test=train_test_split(
    features, df['diagnosis'], test_size=0.2, random_state=0
)
```

```python
knn, dtc = classifier(X_train,X_test,Y_train,Y_test)
plot_confusion_matrix(knn, X_test, Y_test)
plot_confusion_matrix(dtc, X_test, Y_test)
```

# Fazendo validação cruzada

```python
#Agora, vamos utilizar validação cruzada nos algoritmos para verificarmos sua acurácia
def cross_validation(model, x, y, n_folds=10):
    
    kf = StratifiedKFold(n_splits=n_folds)
    
    results_accuracy = []
    
    for train, test in kf.split(x, y):
        data_train, target_train = x[train], y[train]
        data_test, target_test   = x[test], y[test]
        
        model = model.fit(data_train, target_train)
        out = model.predict(data_test)
        
        accuracy = accuracy_score(target_test, out)
        results_accuracy.append(accuracy)
        
    print(f'accuracy: {np.average(results_accuracy)}')
```

```python
x = np.array(features.values)
y = np.array(df['diagnosis'].values)

print("KNN")
cross_validation(knn, x, y)

print("Árvore de Decisão")
cross_validation(dtc, x, y)
```

# Tentando ajustar os Hiperparâmetros para melhorar os resultados

```python
def classifier(X_train,X_test,Y_train,Y_test): 
    #Aplicando KNN
    print(" KNN ")
    ini = time.time() #Inicio do calculo do tempo

    knn=KNeighborsClassifier(n_neighbors=10, leaf_size=20, p=1, n_jobs=1)
    knn.fit(X_train,Y_train)

    end = time.time()
    knn_time = end - ini # Fim do calculo do tempo

    print(classification_report(Y_test,knn.predict(X_test)))
    print()

    #Aplicando Árvore de Decisão
    print("Árvore de Decisão")
    ini = time.time() #Inicio do calculo do tempo

    dtc = tree.DecisionTreeClassifier(criterion="entropy", random_state=1)
    dtc.fit(X_train, Y_train)

    end = time.time()
    dtc_time = end - ini # Fim do calculo do tempo

    print(classification_report(Y_test,dtc.predict(X_test)))

    return knn, dtc
```

```python
knn, dtc = classifier(X_train,X_test,Y_train,Y_test)
plot_confusion_matrix(knn, X_test, Y_test)
plot_confusion_matrix(dtc, X_test, Y_test)
```

```python
x = np.array(features.values)
y = np.array(df['diagnosis'].values)

print("KNN")
cross_validation(knn, x, y)

print("Árvore de Decisão")
cross_validation(dtc, x, y)
```

Podemos notar então uma melhoria considerável no KNN de quase 10% na acurácia da validação cruzada, e de 3% na Árvore de Decisão.
